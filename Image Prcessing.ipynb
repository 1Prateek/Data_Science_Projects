# MODULES USED

NUMPY: It is used to plot image so that it is visible to user.

        fliplr() = fliplr(img) returns img with its columns flipped in the left-right direction (that is, about a vertical axis)

        flipud() = flipud(img) returns img with its rows flipped in the up-down direction (that is, about a horizontal axis).

Matplotlib.pylab:

        imread()-It is used to load an image in the Python program from the specified file.

        imshow()-The imshow() function in pyplot module of matplotlib library is used to display data as an image; i.e. on a 2D          regular raster

SKIMAGE:

        rgb2gray()-The rgb2gray function converts RGB images to grayscale by eliminating the hue and saturation information 
while retaining the luminance. 

```python
import numpy as np
import matplotlib.pylab as plt 
```

```python
from numpy import fliplr, flipud
from skimage.color import rgb2gray
```

This function loads the image from the entered file location

```python
def Load_Image():
    loaction = str(input(" Enter the loaction of the image : ")) # user enters the image path
    image_name = str(input(" Enter the name of the Image : "))   # user enters the image name  
    address = loaction + "\\" + image_name + '.jpg'              # creating address

    image = plt.imread(address)
    return image
```

This function display's the image

```python
def Display_Image():
    img = Load_Image()
    img_plot = plt.imshow(img)                                   # displaying image
    plt.show()
```

This MENU function gives different option to edit the image from selected file location

```python
def Edit_Image():                                                # function for MENU
    choice = int(input('''Press
    1. Horizontal Flip
    2. Vertical Flip
    3. Invert Colors
    4. Convert to Grayscale
    5. Convert to B&W\n'''))

    if (choice == 1):
        img = Load_Image()                                      # original image
        horizontal_flip_image = np.fliplr(img)                  # predefined function to flip image horizontaly
        plt.imshow(img)
        plt.title('Original Image: ')

        plt.imshow(horizontal_flip_image)                       #printing flipped image
        plt.title("Horizontal Flipped Image: ")
        plt.show()
        return horizontal_flip_image

    elif (choice == 2):
        img = Load_Image()                                      # original image
        vertical_flip_image = np.flipud(img)                    # predefined function to flip image vertically
        plt.imshow(img)
        plt.title('Original Image: ')

        plt.imshow(vertical_flip_image)                         #printing flipped image
        plt.title("Vertical Flipped Image: ")
        plt.show()
        return vertical_flip_image

    elif (choice == 3):
        img = Load_Image()                                      # original image
        inverted_image = np.invert(img)                         # predefined function to invert the image
        plt.imshow(img)
        plt.title('Original Image: ')

        plt.imshow(inverted_image)                              #printing inverted image
        plt.title("Inverted Image: ")
        plt.show()
        return inverted_image

    elif (choice == 4):
        img = Load_Image()                                      # original image
        grayscale_image = rgb2gray(img)                         # predefined function to convert the image into grayscale
        plt.imshow(img)
        plt.title('Original Image: ')

        plt.imshow(grayscale_image)                              #printing grayscaled image
        plt.title("Grayscale Image: ")
        plt.show()
        return grayscale_image

    elif (choice == 5):
        img = Load_Image()                                      # original image
        grayscale_image = rgb2gray(img)                         
        plt.imshow(img)
        plt.title('Original Image: ')

        plt.imshow(grayscale_image, cmap=plt.cm.gray)           # converting image to black and white
        plt.title("Black & White Image: ")
        plt.show()
        return grayscale_image

    else:
        print("\n**INVALID CHOICE**\n")
        Edit_Image()
```

```python
while (True):
    ch = (input('''Enter your choice
    1.Load image
    2.Edit image
    3.Save image
    4.Quit\n'''))
    if (ch == '1'):
        Display_Image()
    elif (ch == '2'):
        image = Edit_Image()

    elif (ch == '3'):
        print("--------------------*Saving Image*--------------------")
        Save_Image()
    elif (ch == '4'):
        print("\n\t -----Exiting the MENU-----\t\n")
        break
    else:
        print("\n**INVALID CHOICE**\n")
```
